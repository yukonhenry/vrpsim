```python
import matplotlib.pyplot as plt
import numpy as np
from math import exp
```

```python
x = np.linspace(0, 1300, 2000)
```

```python
# base logistics function
def logistics_func(x, x0, L, k, alpha=None):
    return L/(1+exp(-k*(x-x0)))
```

```python
# more generalized logistics function
def logistics2_func(x, x0, L, k, alpha):
    return L/(1+alpha*exp(-k*(x-x0)))
```

```python
x0 = 500
L = 100
k = 0.01
```

```python
y = [logistics_func(i, x0, L, k) for i in x]
plt.plot(x,y)
```

```python
# derivative of logistics function relative to x value
# derivative form is same for both basic and generalized logistics function
def logistics_deriv_func(func, x, x0, L, k, alpha=None):
    l_val = func(x, x0, L, k, alpha)
    return k*l_val*(1.0-l_val/L)
```

```python
yd = [logistics_deriv_func(logistics_func, i, x0, L, k) for i in x]
plt.plot(x,yd)
```

```python
# parameters for generalized logistics function - note alpha parameter
x0 = 500
L = 100
k = 0.01
alpha = 1.0
```

```python
y = [logistics2_func(i, x0, L, k, alpha) for i in x]
plt.plot(x,y)
```

```python
yd = [logistics_deriv_func(logistics2_func, i, x0, L, k, alpha) for i in x]
plt.plot(x,yd)
```
